```python
import os
import io
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

osPath = os.path.abspath("Tokenizer.ipynb")
inputFile = '/home/dell/Project/MajorProject/AL/code/data/input/record-13.txt'
outputFile = '/home/dell/Project/MajorProject/AL/code/data/output/record-13.txt'

stop_words = set(stopwords.words('english')) 

with open(inputFile) as iFile:
    for line in iFile:
        words = line.split()
        for w in words: 
            if not w in stop_words: 
                print ("Word:", w)
                appendFile = open(outputFile,'a')
                appendFile.write(" "+w) 
appendFile.close() 
        
        
        
```
